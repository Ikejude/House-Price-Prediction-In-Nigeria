````python
# Define relevant functions
# Load dataset into the environment

def load_data(filename: str, sep:str) -> "pd.DataFrame":
    """Returns the loaded dataset as pandas dataframe
    
    Params:
    filename -- str, accepts the dataset and loads as pd.DataFrame
    sep: str, separator mode accepting delimiters of the file
    Return: Pandas dataframe
    """

    # Load the dataset
    df = pd.read_csv(filename, sep=sep)
    
    # Return the dataset
    return df


# Defining functions to handle various tasks

def result_comparison(y_true: np.ndarray, y_pred: np.ndarray) -> pd.DataFrame:
    """Returns new pandas dataframe containing true and predicted labels
    
    Params:
    y_true -- np.ndarray, the initial target variable
    y_pred - np.ndarray, the predicted target variable
    Return: Pandas Dataframe of the true and predicted target variables
    """
    result = pd.DataFrame()
    result["y_true"] = y_true
    result["y_pred"] = y_pred
    result["error_term"] = (y_true - y_pred)

    return result
    

# Defining the evaluation metrics function
def eval_metrics(y_actual: pd.Series, y_predict: pd.Series) -> "str":
    """Returns the evaluation metrics for the model estimator
    
    Params:
    y_true -- pd.Series, pandas series for the true target variable
    y_pred -- pd.Series, pandas series for the predicted target variable
    Return: Scores for the evaluation metrics
    """
    
    mse = round(mean_squared_error(y_true=y_actual, y_pred=y_predict),2)
    rmse = round(mean_squared_error(y_true=y_actual, y_pred=y_predict, squared=False),2)
    mae = round(mean_absolute_error(y_true=y_actual, y_pred=y_predict),2)
    mape = round(mean_absolute_percentage_error(y_true=y_actual, y_pred=y_predict),4)
    r2 = r2_score(y_true=y_actual, y_pred=y_predict)

    ##############Print Result################################
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"MAPE: {mape}")
    print(f"R2: {r2.round(4)}")


# Define function that saves model
def save_model(model, filename: str):
    with open(filename, "wb") as file:
        pickle.dump(model, file)


# Creating functions for numeric variables plots
def num_plot(data: pd.DataFrame, features: list[str]) -> 'plots':
    """Returns histogram for numeric data

    Params:
        data :pd.DataFrame, returns the loaded data
        features: list[str], returns list of features
    """
    plt.figure(figsize=(20,15))
    for i, feat in enumerate(features):
        base_color = sns.color_palette()[0]
        plt.subplot(3,2,i+1)
        axs = sns.histplot(data=data, x=feat, color = base_color, bins=15)
        axs.bar_label(axs.containers[0])
        axs.set_title(f"Histogram for: {feat.title()}")
        axs.set_xlabel(f"{feat.title()}")
        axs.set_ylabel('Frequency')

    plt.tight_layout();


# Plot for categorical variables
# Creating functions for numeric variables plots
def cat_plot(data: pd.DataFrame, features: list[str]) -> 'plots':
    """Returns barchart for categorical data

    Params:
        data :pd.DataFrame, returns the loaded data
        features: list[str], returns list of features
    """
    plt.figure(figsize=(15,10))
    for i, feat in enumerate(features):
        base_color = sns.color_palette()[0]
        plt.subplot(2,2,i+1)
        axs = sns.countplot(data=data, x=df[feat], color=base_color)
        # axs.bar_label(axs.containers[0])
        axs.set_title(f"Bar plot for: {feat.title()}")
        axs.set_xlabel(f"{feat.title()}")
        axs.set_ylabel('Number of apartment')
        plt.xticks(rotation=45)
        var_max = data[feat].value_counts().idxmax()
        var_max_val = data[feat].value_counts()[0]
        var_min = data[feat].value_counts().idxmin()
        var_min_val = data[feat].value_counts()[-1]
        print('{} has the largest number of apartments'.format(var_max),
        'with value',var_max_val)
        print('{} has the smallest number of apartments'.format(var_min),
        'with value',var_min_val)
        print("="*70)

    plt.tight_layout();


# Using heatmap to detect multicollinearity in the dataset. Drop median_house_value, it's the target vector
def heatmap_plot(data: pd.DataFrame) -> "Plot_heatmap":
    """Returns heatmap for numeric variables
    
    Params:
    data -- pd.DataFrame, The loaded dataframe
    Return: Heatmap plot
    """
    corr = data.drop(columns='price', axis=1).corr()
    plt.figure(figsize=[8,4])
    mask = np.triu(np.ones_like(corr, dtype= bool))
    sns.heatmap(data=corr, annot=True, fmt=".2f", cmap='Blues', mask=mask)
    plt.title('Heatmap');


# Making plots of categorical variables against price
def cat_price_plot(data):
    plt.figure(figsize=[20,20])

    ax1 = plt.subplot(3,1,1)
    sns.barplot(data=data, x=data["title"], 
                    y=data["price"]/1e6, ax=ax1)
    ax1.set_title("Barplot of: Title against Price (N Mn)")
    ax1.set_xlabel("Title")
    ax1.set_ylabel("Price(N Mn)")
    plt.xticks(rotation=45)
    var_max = ((df.groupby("title")["price"].mean()/1e6).
               round().sort_values(ascending=False).idxmax())
    var_max_val = ((df.groupby("title")["price"].mean()/1e6).
                   round(2).sort_values(ascending=False).values[0])
    var_min = ((df.groupby("title")["price"].mean()/1e6).
               round().sort_values(ascending=False).idxmin())
    var_min_val = ((df.groupby("title")["price"].mean()/1e6).
                   round(2).sort_values(ascending=False).values[-1])
    print('{} has the largest mean apartment price'.format(var_max),
      'with value N',var_max_val, "Mn")
    print('{} has the smallest mean apartment price'.format(var_min),
      'with value N',var_min_val, 'Mn')
    print("="*70)


    ax2 = plt.subplot(3,1,2)
    sns.barplot(data=data, x=data["state"], 
                    y=data["price"]/1e6, ax=ax2)
    ax2.set_title("Barplot of: State against Price (N Mn)")
    ax2.set_xlabel("State")
    ax2.set_ylabel("Price(N Mn)")
    plt.xticks(rotation=45)
    var_max = ((df.groupby("state")["price"].mean()/1e6).
              round().sort_values(ascending=False).idxmax())
    var_max_val = ((df.groupby("state")["price"].mean()/1e6).
                   round(2).sort_values(ascending=False).values[0])
    var_min = ((df.groupby("state")["price"].mean()/1e6).
               round().sort_values(ascending=False).idxmin())
    var_min_val = ((df.groupby("state")["price"].mean()/1e6).
                   round(2).sort_values(ascending=False).values[-1])
    print('{} has the largest mean apartment price'.format(var_max),
      'with value N',var_max_val, 'Mn')
    print('{} has the smallest mean apartment price'.format(var_min),
      'with value N',var_min_val, "Mn")
    print("="*70)


    ax3 = plt.subplot(3,1,3)
    sns.barplot(data=data, x=data["town"], 
                    y=data["price"]/1e6, ax=ax3)
    ax3.set_title("Barplot of: Town against Price (N Mn)")
    ax3.set_xlabel("Town")
    ax3.set_ylabel("Price(N Mn)")
    plt.xticks(rotation=45)
    var_max = ((df.groupby("town")["price"].mean()/1e6).
              round().sort_values(ascending=False).idxmax())
    var_max_val = ((df.groupby("town")["price"].mean()/1e6).
                   round(2).sort_values(ascending=False).values[0])
    var_min = ((df.groupby("town")["price"].mean()/1e6).
               round().sort_values(ascending=False).idxmin())
    var_min_val = ((df.groupby("town")["price"].mean()/1e6).
                   round(2).sort_values(ascending=False).values[-1])
    print('{} has the largest mean apartment price'.format(var_max),
      'with value N',var_max_val, "Mn")
    print('{} has the smallest mean apartment price'.format(var_min),
      'with value N',var_min_val, "Mn")
    print("="*70)
    
    plt.tight_layout();


# Define function for multivariate data
def multi_plot(data):
    plt.figure(figsize=[20,15])

    plt.subplot(2,3,1)
    sns.scatterplot(data=data, x=num_var["bedrooms"], 
                    y=num_var["bathrooms"], hue=data['title'])


    plt.subplot(2,3,2)
    sns.scatterplot(data=data, x=num_var["parking_space"], y=num_var["price"], 
                    hue=data['state'])

    plt.subplot(2,3,3)
    sns.scatterplot(data=data, x=num_var["bedrooms"], y=num_var["price"], 
                    hue=data['title'])


    plt.subplot(2,3,4)
    sns.scatterplot(data=data, x=num_var["bedrooms"], y=num_var["price"], 
                    hue=data['state'])


    plt.subplot(2,3,5)
    sns.scatterplot(data=data, x=num_var["bathrooms"], 
                    y=data["parking_space"], hue=data['title'])
    
    plt.subplot(2,3,6)
    sns.scatterplot(data=data, x=num_var["bedrooms"], 
                    y=data["parking_space"], hue=data['state'])

    plt.tight_layout();


# Encoding categorical variables
def cat_encoding(data: pd.DataFrame, features: list[str]) -> "pd.DataFrame":
    """Returns pandas dataframe with the categorical features transformed
    
    Params:
    data -- pd.DataFrame, the input dataframe
    features -- list[np.ndarray], contains list of pd.Series object
    Return: returns dataframe with transformed categorical features
    """
    cat_transformer = {}
    feat_list = []
    for feat in features:
        var_cat = data[feat].value_counts().sort_index()
        for idx, col in enumerate(var_cat.index):
            cat_transformer[col] = idx
        cat_transformer
        data[feat] = data[feat].map(cat_transformer)
        # feat_list.append(cat_transformer)
    # return feat_list
    return data


# Create prediction function
def make_prediction(
        bedrooms, 
        bathrooms, 
        toilets, 
        title,
        town, 
        state
        ):
    
    """Returns the predicted apartment in Nigeria
    Params:
    bedrooms: pd.Series[int], accepts the number of bedrooms
    bathrooms:pd.Series[int], accepts the number of bathrooms
    toilets:pd.Series[int], accepts the number of toilets
    title:pd.Series[int], accepts the number of title
    town:pd.Series[int], accepts the number of town
    state:pd.Series[int], accepts the number of state

    Returns:
        pd.Series(int): Predicted apartment price
    """

    data = {
        'bedrooms': bedrooms,
        'bathrooms': bathrooms,
        'toilets': toilets,
        'title': title,
        'town': town,
        'state': state,
    }
    df = pd.DataFrame(data=data, index=[0])
    prediction = model_rfc.predict(df).round(2)[0]
    return f"Predicted apartment price: N{prediction}"
````
